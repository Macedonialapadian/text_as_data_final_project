# Day 3: Further Pre-Training a Model

Sometimes, you might have some especilized text (e.g., legislative speeches from the Brazilian legislature, tweets during a indigenous protest in Ecuador) that is not likely to be properly 'understood' by any pre-trained model: either because the training language is not the same (e.g,. technical language), or because the (high-information) tokens used in the target corpus are not likely to appear in the pre-trained data (e.g., local slang or phrasing), or because the training data is older than a specific event (e.g., COVID). We can improve an already pre-trained model by **further** pre-training it. For this, you will need (a lot) more unstructured data, and (a lot of) computational resources. For an example of a supervised Transformers-based machine learning classification approach from start to finish, that includes further pre-training a model, as well as the theoretical foundations to do so, you can check the following [working paper and appendix](https://github.com/svallejovera/iesp-uerj/blob/main/racism_nlp_ddg_jt_svv.pdf) (please, do not circulate it). Here is the annotated code on how to do it.

## 1. Adding tokens

The first step is optional, but often desirable. We can add tokens that do not appear in the pre-trained model's dictionary, or we can modify the embeddings of current tokens to more closely resemble the 'meaning' of how they appear in our target corpus. Let's further pre-train a BERT model. First, we will add some tokens that do not originally appear on its dictionary.

```python
# Import the usual libraries
from transformers import BertTokenizer
from transformers import BertForSequenceClassification
from transformers import BertConfig, BertModel, BertForMaskedLM
from transformers import LineByLineTextDataset, DataCollatorForLanguageModeling
from transformers import Trainer, TrainingArguments
import torch
import matplotlib.pyplot as plt
import pandas as pd
```

For pre-training, we will import the masked language model function `BertForMaskedLM` since, as might remember from the lecture, this is how these models are trained.

```python
maskedlm_model = BertForMaskedLM.from_pretrained("bert-base-uncased")
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased", do_lower_case=True)
```

Maybe we want to classify a corpus that heavily includes text about COVID or the pandemic. COVID is unlikely to feature heavily on the pre-training data, or might no feature at all:

```python
print(tokenizer.tokenize('COVID'))
print(tokenizer.tokenize('coronavirus'))
```

The words are not in the data, but in our target corpus these might be high-information tokens. Let's add them to the model's dictionary (i.e., to the tokenizer).

```python
print (len(tokenizer))
tokenizer.add_tokens(["covid"])
tokenizer.add_tokens(["coronavirus"])
print (len(tokenizer))
```

We have added them. Since the model has not been trained yet with these tokens, they are assigned a random embedding (meaningless). Now we have two options: 1) we can further pre-train the model and let it update the embedding with new, more accurate values as it sees more of the tokens in the new data we will use, or 2) we can 'help' the model a bit by assigning an embedding of our own to the new tokens. Let's do the latter.


```python
# FOR CORONAVIRUS:
maskedlm_model.resize_token_embeddings(len(tokenizer))
random_vector = maskedlm_model.get_input_embeddings().weight[-1].detach().numpy() # last one is coronavirus random vector


## This is a visual representation of the random embedding assigned to coronavirus:
plt.title("Randomly Initialized Vector")
plt.hist(random_vector, bins=50)
plt.show()

## From our understanding of the world, we know that covid / coronavirus is a respiratory virus.
## One way to 'help' the model is by assigning the average (why average? why not!) of the embedding for 'respiratory' and 'virus'
virus_id = tokenizer.convert_tokens_to_ids("virus")
respiratory_id = tokenizer.convert_tokens_to_ids("respiratory")

virus_embedding = maskedlm_model.get_input_embeddings().weight[virus_id]
respiratory_embedding = maskedlm_model.get_input_embeddings().weight[respiratory_id]

mean_embedding = torch.mean(torch.stack([virus_embedding, respiratory_embedding]), dim=0)

maskedlm_model.get_input_embeddings().weight[-1].data[:] = mean_embedding
mean_vector = maskedlm_model.get_input_embeddings().weight[-1].detach().numpy()

plt.title("Custom Initialized Vector")
plt.hist(mean_vector, bins=50)
plt.show()

## Note that these embeddings will change as we further pre-train the model. These is just the starting point for our new token.

# For COVID:
maskedlm_model.get_input_embeddings().weight[-2].data[:] = mean_embedding
```

## 2. Further Pre-Training a Model

Ok, now that we have all ready, we can further pre-train our BERT model using the unstructured raw data we have.

```python
# Load the raw data
dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path="abstracts.txt",
    block_size=32,
)

# Mask the data, and choose the masking rate
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=True, mlm_probability=0.15 # <- 15%... why? why not? are there better ways... maybe...
)
```

If you want to explore different masking strategies you can check out [this paper](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0318421) for some alternatives, and [this code](https://github.com/svallejovera/masking_plosone/blob/main/README.md) to implement them.

```python
# Determine the hyperparameters (usually, similar to how the original model was trained)
training_args = TrainingArguments(
    output_dir="./",
    overwrite_output_dir=True,
    learning_rate=5e-06,
    num_train_epochs=1,
    per_gpu_train_batch_size=16,
    save_steps=10000,
    #save_total_limit=2,
)

# Set up trainer
trainer = Trainer(
    model=maskedlm_model,
    args=training_args,
    data_collator=data_collator,
    train_dataset= dataset
)
```

I won't run the next lines becuase it will take a **long** time... like a **looooonnnggg** time...

```python
trainer.train()
```

```python
trainer.save_model("covid_trained_bert")
```

```python
tokenizer.save_pretrained("covid_trained_bert_tok")
```
